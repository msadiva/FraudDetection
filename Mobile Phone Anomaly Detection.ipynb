```python
import os
import pandas as pd
```

```python
df = pd.read_csv("Dataset_Cell_Phones_Model_Brand.csv")
```

```python
drop_columns = [
    'objectId', 'createdAt', 'updatedAt', 'ACL',
    'Cell_Phone_Models', 'Cell_Phone_Brands',
    'field13', 'field38'
]

# Drop the columns safely (if they exist in the DataFrame)
df = df.drop(columns=[col for col in drop_columns if col in df.columns])

# Check the remaining columns
print("Remaining columns:", df.columns.tolist())
```

```python
pd.set_option('display.max_columns', 500)

df.head()
```

```python
import re
import numpy as np

# Copy df to avoid modifying original
df_cleaned = df.copy()

# ----------- 1. Normalize Binary Columns -----------

binary_cols = ['FourG', 'ThreeG', 'TwoG', 'WLAN', 'GPS', 'GPRS', 'NFC', 'Radio', 'Loud_speaker', 'Audio_jack']
for col in binary_cols:
    df_cleaned[col] = df_cleaned[col].astype(str).str.lower().str.contains('yes|true|hotspot').astype(int)

# ----------- 2. Extract Numeric Values -----------

def extract_number(text):
    if pd.isnull(text):
        return np.nan
    match = re.search(r'(\d+(\.\d+)?)', text)
    return float(match.group(1)) if match else np.nan

# Columns with sizes or memory values
numeric_cols = ['Internal_memory', 'Battery', 'RAM', 'Display_size', 'Display_resolution']
for col in numeric_cols:
    df_cleaned[col + '_num'] = df_cleaned[col].apply(extract_number)

# Example: convert RAM to GB
df_cleaned['RAM_GB'] = df_cleaned['RAM'].str.extract(r'(\d+)').astype(float)

# ----------- 3. Split/Encode Complex Columns -----------

# SIM type
df_cleaned['SIM_type'] = df_cleaned['SIM'].fillna('Unknown').astype(str)

# Operating System
df_cleaned['OS'] = df_cleaned['Operating_System'].fillna('Unknown').astype(str)

# Chipset and CPU – fill missing or label encode later
df_cleaned['Chipset'] = df_cleaned['Chipset'].fillna('Unknown')

# Sensors – count number of sensors
df_cleaned['Sensor_count'] = df_cleaned['Sensors'].fillna('').apply(lambda x: len(str(x).split('|')) if x else 0)

# ----------- 4. Drop or Impute Missing -----------

# Drop rows with too much missing info
df_cleaned.dropna(thresh=int(df_cleaned.shape[1]*0.5), inplace=True)

# Optional: fill remaining missing
df_cleaned.fillna('Unknown', inplace=True)

# Preview cleaned data
print(df_cleaned.head())
```

```python
drop_redundant_cols = [
    'Internal_memory', 'Battery', 'RAM', 'Display_size', 'Display_resolution'
]
df_cleaned.drop(columns=drop_redundant_cols, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder


categorical_cols = ['Brand', 'CPU', 'Chipset', 'SIM_type', 'OS']
le = LabelEncoder()

for col in categorical_cols:
    df_cleaned[col] = le.fit_transform(df_cleaned[col].astype(str))
```

```python
df_cleaned.head()
```

```python
from sklearn.ensemble import IsolationForest

# Choose only numeric columns for modeling
numeric_df = df_cleaned.select_dtypes(include=[np.number])

model = IsolationForest(contamination=0.05, random_state=42)
model.fit(numeric_df)

# Predict anomalies (-1 for anomaly, 1 for normal)
df_cleaned['anomaly'] = model.predict(numeric_df)

# Filter anomalies
anomalies = df_cleaned[df_cleaned['anomaly'] == -1]
print("Number of anomalies found:", len(anomalies))
```
